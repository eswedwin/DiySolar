Check for the optimum fixed angles for 0,1,2 axi of movement with constant or TOU tariffs.
First enter all the constants in the program below (Latitude -south, +north).
doy 1 = 1 January, 365 = 31 December, -10 = 21 December, 182 - 10 = 21 June (soltices)

```python
LATITUDE_deg = -25.79    # enter your latitude here in decimal degrees (+N, -S)
INTERVAL_hours = 0.25    # calculate every 15 minutes
RANGE_hours = 12.0       # include both ends

import math as m
import numpy as np
%pylab inline


st = ((RANGE_hours / INTERVAL_hours) // 2) * INTERVAL_hours
hours = arange(-st, st + INTERVAL_hours, INTERVAL_hours, dtype=float)
hours_rad = hours * (np.pi / 12)
own_latitude_rad = np.radians(LATITUDE_deg)
NOON = size(hours) // 2

def sun_latitude_rad(doy):
    return (np.radians(-23.44 * np.cos(np.pi * (doy + 10) / 182.5)))

def local_sun_angle(doy):
    zenith_angle_rad = \
        np.outer((np.sin(own_latitude_rad) * np.sin(sun_latitude_rad(doy))).T,ones(size(hours_rad))) \
      + np.outer((np.cos(own_latitude_rad) * np.cos(sun_latitude_rad(doy))).T,np.cos(hours_rad))
    zenith_angle_rad = np.arccos(zenith_angle_rad * (zenith_angle_rad >= 0))
    return(zenith_angle_rad)
    
def norm_air_mass(doy):    
    air_mass = 1 / np.cos(local_sun_angle(doy)) - 10  # approximate AirMass rays have to pass through
    # - no need to comp for curvature        # MAM = AM * (1 - 601 * (1 / np.cos(zar/600) - 1))
    air_mass = air_mass*(air_mass < 0) + 10  # 1 atmosphere height at 0 angle, limit to max 10
    return(air_mass)

def cos_theta_i_panel(doy, tilt_ns_rad, tilt_ew_rad):
    angle_rad = np.outer((np.sin(own_latitude_rad + tilt_ns_rad) * np.sin(sun_latitude_rad(doy) +tilt_ew_rad)).T \
              , ones(size(hours_rad))) \
              + np.outer((np.cos(own_latitude_rad + tilt_ns_rad) * np.cos(sun_latitude_rad(doy) +tilt_ew_rad)).T \
              , np.cos(hours_rad))
    angle_rad *= (angle_rad > 0)          # no negative, force zero
    angle_rad += 0.001 * (angle_rad <= 0) # no zero, force small positive
    angle_rad -= 0.00000001               # slightly less than 1
    return(angle_rad)                     # arccos here for real angle 
    
def norm_sun_vector(doy, tilt_ns_rad, tilt_ew_rad):
    panel_lat = sun_latitude_rad(doy) - own_latitude_rad - tilt_ns_rad
    # local_sun_vector (normalised) used for calculating energy for tilted and rotating panels
    local_sun_vector = zeros((size(doy),size(hours_rad),3),dtype=float)   # (x,y,z) x rot, y north, z = zenith
    local_sun_vector[:, :, 0] = np.outer(ones(size(doy)).T,   np.sin(hours_rad + tilt_ew_rad)) # x
    local_sun_vector[:, :, 1] = np.outer(np.sin(panel_lat).T, np.cos(hours_rad + tilt_ew_rad)) # y
    local_sun_vector[:, :, 2] = np.outer(np.cos(panel_lat).T, np.cos(hours_rad + tilt_ew_rad)) # z
    return (local_sun_vector)

def fresnel_refl(n1,c2,n2,c1):
    return (((n1*c1 - n2*c2)/(n1*c1 + n2*c2))**2)

def calc_tou_tariff_vec(hours, TARIFF):
    peak = 1 * (hours < -2)
    peak = peak + 1 * (hours >= 6)
    peak = peak * (hours >=-5)
    normal = 1 * (hours < -5)
    normal = normal + 1 * (hours >= -2)
    normal = normal * (hours < 6)          # no offpeak during sun-hours on workdays
    tariff_vec = TARIFF[0] * peak
    tariff_vec = tariff_vec + TARIFF[1] * normal
    return (tariff_vec)

```

```python
TILT_NS_deg = 32.0                 # for same peak day-energy at soltices and 17% more at equinoxes
TILT_EW_deg = 00.0                 # for more energy during peak for TOU
AXI = 0                            # change to 1 for e-w rotation, and 2 for addition n-s rot
MODULE_eff = 0.17                  # converts upto 17% of incedent light into electricity @ 25oC
TEMP_grad = 0.004                  # derate performance for each degree above 45oC
DAYS_SIZE = 52                     # 4.36 samples / month
INSOLATION = 2000                  # kWh/m^2 as measured over 10 years here and averaged

days = arange(-10, 365 - 10, 365 // DAYS_SIZE, dtype = float)
tilt_ns_rad = np.radians(TILT_NS_deg)
tilt_ew_rad = np.radians(TILT_EW_deg)
if (AXI == 2):
    tilt_ns_rad = sun_latitude_rad(days) - own_latitude_rad
if (AXI >= 1):
    tilt_ew_rad = -hours_rad

AM = norm_air_mass(days)                                # mostly absorbtion and raleigh scattering
NSV = norm_sun_vector(days, tilt_ns_rad, tilt_ew_rad)
pepend_sun_power_kw = 1.367 - (AM + 0.5) * 0.367 / 2.0  # per square m (0.5 = reflection of atmos)
pepend_sun_power_kw = pepend_sun_power_kw * (pepend_sun_power_kw > 0)  # force positive only
pepend_sun_power_kw = pepend_sun_power_kw * MODULE_eff  # max power from panel if follow sun
norm_panel_power_kw = pepend_sun_power_kw * NSV[:, :, 2]   # norm_pepend_area = z
#print(AM[NOON], pepend_sun_power_kw[NOON])

if (1):
    scale = INTERVAL_hours * 365 / size(days)
    panel_energy = np.sum(norm_panel_power_kw) * scale
    print(np.sum(pepend_sun_power_kw) * scale, '-',\
          np.sum(pepend_sun_power_kw - norm_panel_power_kw) * scale, '=',\
          panel_energy, 'kWh / m^2')
    print('with 303 / 365 sun days from above:', (panel_energy * 5 / 6).round(1))
    print('Max panel energy from insolation:  ', INSOLATION * MODULE_eff)
    plot(hours, pepend_sun_power_kw[0], '-r')
    plot(hours, norm_panel_power_kw[0], '-b')
    plot(hours, pepend_sun_power_kw[0]-norm_panel_power_kw[0], '-y')
    plot(hours, pepend_sun_power_kw[DAYS_SIZE // 2], '-r')
    plot(hours, norm_panel_power_kw[DAYS_SIZE // 2], '-b')
    plot(hours, pepend_sun_power_kw[DAYS_SIZE // 2]-norm_panel_power_kw[DAYS_SIZE // 2], '-y')
    xlabel('Time relative to noon in hours')
    ylabel('pependicular (r)/norm sun power (b) kw')
    show()
```

Reflection:
-----------
Default (theta_i = 0) reflection already calculated into the panel spec Wp.

Use normalised reflection/transmission to calculate power at other incedence angles.

Air (No=1.0), silicon (Ns=4.3) interface (Reff = 0.388, Teff = 0.612) (theta_i = 0)

with optimal anti-reflective coating (Na=2.07): Reff = 0.229, Teff = 0.771 (theta_i = 0)

snell law --> n1 * sin (theta_i) = n2 * sin (theta_t) (theta measured from norm on interface)

fresnel reflection for non-complex refractive indices (excludes / approximates semiconductors):

Rp = ((n1 * cos(theta_t) - n2 * cos(theta_i)) / (n1 * cos(theta_t) + n2 * cos(theta_i))) ^ 2,

Rs = ((n1 * cos(theta_i) - n2 * cos(theta_t)) / (n1 * cos(theta_i) + n2 * cos(theta_t))) ^ 2,

R = (Rp + Rs) / 2,

T = 1 - R

```python
N_air = 1
N_antir = 2.07
N_silicon = 4.3
if (tilt_ew_rad.size == 1):
    cos_theta_i = (cos_theta_i_panel(days, tilt_ns_rad, tilt_ew_rad))
else:
    cos_theta_i = (cos_theta_i_panel(days, tilt_ns_rad, 0))
cos_theta_t = (np.arcsin(N_air * np.sin(np.arccos(cos_theta_i)) / N_antir))
cos_theta_f = np.cos(np.arcsin(N_antir * np.sin(cos_theta_t) / N_silicon))
cos_theta_t = np.cos(cos_theta_t)

T_norm = 1 - (fresnel_refl(N_air, cos_theta_t, N_antir, cos_theta_i) \
  +  fresnel_refl(N_air, cos_theta_i, N_antir, cos_theta_t) \
  +  fresnel_refl(N_antir, cos_theta_f, N_silicon, cos_theta_t) \
  +  fresnel_refl(N_antir, cos_theta_t, N_silicon, cos_theta_f) \
  ) / 2                                       # calcs to 0.75 iso 0.77 at theta_i=0

#print ((cos_theta_i)[:,NOON],T_norm[:,NOON], np.max(T_norm))
T_norm *= (T_norm > 0)                        # force not negative
T_norm += 0.01 * (T_norm <= 0)                # force not zero
T_norm = (T_norm / np.max(T_norm)).round(2)   # Normalise to the max_transmission of panel

transm_panel_power_kw = norm_panel_power_kw * T_norm

if (1):
    scale = INTERVAL_hours * 365 / size(days)
    print(np.sum(pepend_sun_power_kw) * scale, '-',\
          np.sum(pepend_sun_power_kw - transm_panel_power_kw) * scale, '=',\
          np.sum(transm_panel_power_kw) * scale, 'kWh/m^2')
    plot(hours, pepend_sun_power_kw[0], '-r')
    plot(hours, transm_panel_power_kw[0], '-b')
    plot(hours, pepend_sun_power_kw[0]-transm_panel_power_kw[0], '-k') # oblique angle smaller area
    plot(hours, transm_panel_power_kw[0] * (1 / T_norm[0] - 1), '-k')  # reflection losses
    plot(hours, pepend_sun_power_kw[DAYS_SIZE // 2], '-m')
    plot(hours, transm_panel_power_kw[DAYS_SIZE // 2], '-c')
    plot(hours, pepend_sun_power_kw[DAYS_SIZE // 2] 
            - transm_panel_power_kw[DAYS_SIZE // 2], '-y') # oblique angle smaller area
    plot(hours, norm_panel_power_kw[DAYS_SIZE // 2] 
                 * (1 / T_norm[DAYS_SIZE // 2] - 1), '-g')  # reflection losses
    xlabel('Time relative to noon in hours')
    ylabel('pependicular (r)/norm sun power (b) kw')
    show()
```

```python
WINTER_TARIFF = WINTER_TARIFF_11KV_2014 = [3.328, 0.870, 0.466]   # 11KV (2014) no peak on saturdays
SUMMER_TARIFF = SUMMER_TARIFF_11KV_2014 = [0.933, 0.573, 0.402]   # 11KV (2014) only offpeak on sundays
NORMAL_TARIFF_home_2014 = 1.4723                  # home (2014) max marginal cost / kWh at all times
INCREASE_2015 = 1.000 + 0.127                     # 12.7% announced 3 oct 2014

for i in range(3):
    WINTER_TARIFF[i] = WINTER_TARIFF_11KV_2014[i] * 1.60  # home 60% more expensive than 11KV energy
    SUMMER_TARIFF[i] = SUMMER_TARIFF_11KV_2014[i] * 1.60  # 11KV includes KVA charge @ R119.00/kVA
    WINTER_TARIFF[i] *= INCREASE_2015
    SUMMER_TARIFF[i] *= INCREASE_2015
NORMAL_TARIFF = NORMAL_TARIFF_home_2014 * INCREASE_2015   # 8% minimum + 4.7% special increase

tariff_w_r_kwh = calc_tou_tariff_vec(hours, WINTER_TARIFF)
tariff_s_r_kwh = calc_tou_tariff_vec(hours, SUMMER_TARIFF)
tariff_o_r_kwh = SUMMER_TARIFF[2] * (hours < 10)       # offpeak tariff all hours = sundays
tariff_n_r_kwh = NORMAL_TARIFF * (hours < 10)          # max for more than 650kwh usage a month

scale_day = INTERVAL_hours                             # back to hours
w_sta = DAYS_SIZE // 2 - DAYS_SIZE // 12               # mid year - one month
w_per = DAYS_SIZE // 4                                 # 3 months
# savings for the all calculated days when multiplied with scale for hours
saving_d_t = zeros(transm_panel_power_kw.shape[0], dtype = float)
saving_d_t[w_sta:w_sta+w_per] = np.inner(transm_panel_power_kw[w_sta:w_sta+w_per], tariff_w_r_kwh)
saving_d_t[:w_sta] = np.inner(transm_panel_power_kw[:w_sta], tariff_s_r_kwh)
saving_d_t[w_sta+w_per:] = np.inner(transm_panel_power_kw[w_sta+w_per:], tariff_s_r_kwh)
saving_d_t *= scale_day

saving_d_o = np.inner(transm_panel_power_kw, tariff_o_r_kwh) * scale_day
saving_d_n = np.inner(transm_panel_power_kw, tariff_n_r_kwh) * scale_day

if (1):
    print('average daily saving R tou,nor', np.average(saving_d_t), ',', np.average(saving_d_n))
    plot(days + 10, saving_d_t, '-r')
    plot(days + 10, saving_d_o, '-y')
    plot(days + 10, saving_d_n, '-b')
    xlabel('day of year (all sunny)')
    ylabel('max savings for the day in R')
    ylim(0,)
    show()
```

Calculate and print yearly savings with different tariffs, project cost and payback period with sunny days as in pretoria

```python
YEAR_DAYS = 365
WINTER_DAYS = 92                        # june, july & august, all sun subtract weekends and holidays
YEAR_SUNNY_DAYS = 304                   # 61 summer days completely cloudy with no power (not true)

PANEL_COST = 12.00                      # / Wp = $1.10 * R10.70 / Wp - small scale (large * 0.75 to 0.50)
BOS = 6.00                              # / Wp balance of system for diy

# total savings for the all days in range when multiplied with scale for the other days
scale_year = YEAR_DAYS / size(days)                     # back to all days in year
saving_w = (np.sum(saving_d_t[w_sta:w_sta+w_per]) * scale_year).round(2) # split for sunny_days
saving_s = (np.sum(saving_d_t) * scale_year).round(2) - saving_w  # cloudy days only in summer
saving_o = (np.sum(saving_d_o) * scale_year).round(2)
saving_n = (np.sum(saving_d_n) * scale_year).round(2)   # all sun
# assume 64(w) + 136(s) = 200 sunny work days, rest weekend or offpeak/cloudy
adjusted_tou =  (saving_w * 64 / WINTER_DAYS \
               + saving_s * 136 / (YEAR_DAYS - WINTER_DAYS) \
               + saving_o * 165 / YEAR_DAYS \
               + saving_o * 52 / 2 / YEAR_DAYS).round(2) # add some sunny saturdays = standard tariff

adjusted_nor = (saving_n * YEAR_SUNNY_DAYS / YEAR_DAYS).round(2) # with usage > generated on weekends
                                                                 # other 61 days = 0
total_cost = (PANEL_COST + BOS) * MODULE_eff * 1000

print ('Saving tou all (w(3),s(9)):', saving_w, '+', saving_s, '=', saving_w + saving_s, 'R/m^2')
print ('Saving off-pk year all sun:', saving_o, 'R/m^2')
print ('Saving normal year all sun:', saving_n, 'R/m^2\n')
print ('First year saving adjusted sunny days for tou, nor:', adjusted_tou, ',', adjusted_nor, 'R/m^2')
print ('Total installed cost / m^2 = R', total_cost)
print ('Payback years for tou, nor:',  (total_cost / adjusted_tou).round(1), ',', \
                                       (total_cost / adjusted_nor).round(1))
```

Financial calculations over life of system.
-------------------------------------------
Calculate:
NPV (nett present value) with constant rate for capital cost (return).  This is much more realistic, relative to return on capital invested, with costs, present value and escalation calculations.

With estimated electricity tariff escalation, inflation and detoriation of PV conversion efficiency.

Assume system is scrapped and removed after 30 years with removal cost == profit from scrap. Thus no cost or return. 

```python
CALC_YEARS  = 30            # life of PV panels normally set to 25 - 30 Years
TARIFF_ESCAL = 0.08         # electricity increase at minimum 8% per year, thus 1.5% above inflation
INFLATION    = 0.065        # current RSA
PRIME_INT_RATE = 0.0925     # current RSA
PV_EFF_25Y   = 0.8          # 80% after 25 years because of accumelated radiation and thermal damage
MAINT_COST   = 0.025        # 2.5% pa. escalated cleaning/painting over long term when no batteries
INSUR_COST   = 0.02         # 2.0% pa. esc. if needed for hail damage/ theft, both increase with inflation
REQUIRED_RETURN = 0.12      # rate of return (coc) for NPV calc (prime + 2.75%)

# from above use adjusted_tou, adjusted_nor as returns and total_cost as initial capital cost

years = arange(0, CALC_YEARS + 1, 1)        # calculation period
life_expense = total_cost * (MAINT_COST) * (1 + INFLATION) ** years
life_expense[0] = 0                         # maintenance only after first year
life_expense += total_cost * (INSUR_COST) * (1 + INFLATION) ** years # insurance paid forwards
pv_eff_dec = 1 - PV_EFF_25Y ** (1 / 25)     # (1 - pv_eff) ^ 25 = 0.8 ==> pv_eff = 1 - 0.8 ** (1/25) = 0.0089
return_ajustment = ((1 - pv_eff_dec) * (1 + TARIFF_ESCAL)) ** years
return_ajustment[0] = 0                     # return only after first year
return_tou = adjusted_tou * return_ajustment - life_expense 
return_nor = adjusted_nor * return_ajustment - life_expense

# calculate NPV (nett present value) with fixed rate of return
# npv_tou_1 = -total_cost + np.npv(REQUIRED_RETURN, return_tou)      # gives single value (the same as below)
# npv_nor_1 = -total_cost + np.npv(REQUIRED_RETURN, return_nor)      # gives single value
# np.nper(PRIME_INT_RATE + 0.001, -npv_nor, total_cost)

pv_tou = return_tou / (1 + REQUIRED_RETURN) ** years 
pv_nor = return_nor / (1 + REQUIRED_RETURN) ** years 
npv_tou = -total_cost + sum(pv_tou).round(2)
npv_nor = -total_cost + sum(pv_nor).round(2)
npv_tt = np.ma.cumsum(pv_tou)
npv_tt -= total_cost 
npv_nt = np.ma.cumsum(pv_nor)
npv_nt -= total_cost 

if (1):
    #print ('NPV/COST nor =', npv_nor*100// total_cost, npv_nor_1*100//total_cost)
    print ('NPV/Cost tou, nor =', npv_tou * 100 // total_cost,'%', npv_nor * 100 // total_cost,'%')
    print ('excess return normalised for tou, nor =', \
           (npv_tou * 100 // total_cost / CALC_YEARS).round(1),'% pa.', \
           (npv_nor * 100 // total_cost / CALC_YEARS).round(1),'% pa.')
    #plot(years, return_nor[1:], '-r')
    plot(years[1:], pv_nor[1:], '-b')
    #plot(years, return_tou, '-g')
    plot(years[1:], pv_tou[1:], '-y')
    plot(years, npv_nt / 10, '-c')    # divided 10 below WATCH!
    plot(years, npv_tt / 10, '-m')    # divided 10 below WATCH!
    xlim(0,)
    ylim(0, 500)
    legend (['pv_nor_P', 'pv_tou_P', 'npv_nor/10', 'npv_tou/10'])
    show()


```
